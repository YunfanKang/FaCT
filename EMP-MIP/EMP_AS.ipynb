```python
from gurobipy import *
import numpy as np
import math
from random import randrange
```

```python
g = 3
n = int(math.pow(g, 2)) + 1
q = n - 1 
sumLow = 120
avgLow = 25
avgHigh = 45
```

```python
A = list(range(1,n))
```

```python
T = tuplelist()
for i in range(1, n):
    for j in range (1, n):
        T.append((i, j))
```

```python
X = tuplelist()
for i in range(1, n):
    for k in range (1, n):
        for c in range (0, q):
            X.append((i, k, c))
```

```python
#L = [30, 25, 31, 28, 32, 30, 35, 27, 33]
#Y = [350.2, 400.5, 430.8, 490.4, 410.9, 450.4, 560.1, 500.7, 498.6]
L=[]
Y=[]
A = []
for i in range(1,n):
    L.append(randrange(20,40))
    Y.append(10 * randrange(30,60))
    A.append(randrange(10, 60))
    
```

```python
L
```

```python
w = []
for i in range(1,n):
    wi = []
    for j in range(1,n):
        if(abs(i-j) == 1 or abs(i-j) == g):
            wi.append(1)
        else:
            wi.append(0)
    w.append(wi)
```

```python
w
```

```python
d = dict()
for i in range(1, n):
    for j in range(1, n):
        #d[i][j] = abs(Y[i] - Y[j])
        
        if(j > i):
            d[(i, j)] = abs(Y[i-1] - Y[j-1])
        else:
            d[(i, j)] = 0

N = []
for i in range(1, n):
    Ni = []
    for j in range(1, n):
        if(w[i-1][j-1]) == 1:
            Ni.append(j)
    N.append(Ni)

            

```

```python
N
```

```python
d
```

```python
h = 1
totalDis = 0;
for i in range(1, n):
    for j in range(i+1, n):
        totalDis += d[(i,j)]
        #print(i, " ", j, " ", d[i][j], " ", totalDis)
        
h += math.floor(math.log(totalDis))
```

```python
h
```

```python
m = Model('EMP')
```

```python
x = m.addVars(X, vtype=GRB.BINARY, name="x")
t = m.addVars(T, vtype=GRB.BINARY, name="t")
```

```python
#objective function (not considering minimizing the no. unassigned areas)
m.setObjective(-1* quicksum(x[i,k,0] for i in range(1,n) for k in range(1,n)) * pow(10, h) + quicksum(t[i,j] * d[i,j] for i in range (1,n) for j in range(1,n) if j>i))
```

```python
regionInit = m.addConstrs((x.sum('*', k, 0) <= 1 for k in range(1, n)), 'regionInit')
```

```python
noDup = m.addConstrs((x.sum(i, '*', '*') == 1 for i in range (1, n)), 'noDup')
```

```python
#neighbor = m.addConstrs((x[i, k, c] <= x.sum('*', k, c-1) for i in range (1, n) for k in range(1, n) for c in range(1, 1)), "neighbor")
neighbor = m.addConstrs((x[i, k, c] <= quicksum(x[j, k, c-1] for j in N[i-1]) for k in range(1,n) for c in range(1,q) for i in range(1,n)), name='contiguity')
```

```python
sumation = m.addConstrs((quicksum(x[i,k,c] * L[i-1] for i in range(1,n) for c in range(0,q)) >= quicksum(x[i, k, 0] * sumLow for i in range(1,n)) for k in range(1,n)),  name = "sum" )
```

```python
avgLow = m.addConstrs((quicksum(x[i,k,c] * A[i-1] for i in range(1,n) for c in range(0,q)) >= quicksum(x[i, k, 0] * avgLow for i in range(1,n)) * quicksum(x[i,k,c] for i in range(1,n) for c in range(0,q)) for k in range(1,n)),  name = "avgLow" )
```

```python
aveHigh = m.addConstrs((quicksum(x[i,k,c] * A[i-1] for i in range(1,n) for c in range(0,q)) <= quicksum(x[i, k, 0] * avgHigh for i in range(1,n)) * quicksum(x[i,k,c] for i in range(1,n) for c in range(0,q)) for k in range(1,n)),  name = "avgHigh" )
```

```python
tij = m.addConstrs((quicksum(x[i,k,c]+x[j,k,c] for c in range(0,q)) -1 <= t[i,j] for i in range(1,n) for j in range(1,n) if i < j for k in range(1,n)), name = "tij")
```

```python
m.write('MP.lp')
```

```python

m.optimize()
```
